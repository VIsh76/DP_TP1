### 3 Visualize the weights

```python
W = model.get_weights()
def Visu_Weights(W):
    im = W[0].reshape(72,72,3)
    c = 255/(np.max(im)-np.min(im))
    im = (c*(im - np.min(im))).astype(int)
    f, axs = plt.subplots(1,3,figsize=(20,5))
    axs[0].imshow(im[:,:,0], cmap="gray")
    axs[1].imshow(im[:,:,1], cmap="gray")
    axs[2].imshow(im[:,:,2], cmap="gray")

Visu_Weights(W)
plt.show();
```

Each weight has to detect a shape.

The first column detects the sides of a square

The second tries to detect the round of a circle.

The last one detects the angles and sides of a triangles.

```python
from keras.layers import Conv2D, MaxPooling2D, Flatten
x_train = X_train.reshape(X_train.shape[0], 72,72, 1)
y_train = keras.utils.to_categorical(Y_train, 3)

x_test = X_test.reshape((X_test.shape[0], 72,72, 1))
y_test = keras.utils.to_categorical(Y_test, 3)

# Model
model2 = Sequential()
model2.add(Conv2D(16, kernel_size=(5,5), activation="relu",input_shape=X_train.shape[1:]))
model2.add(MaxPooling2D(pool_size=(3,3)))
model2.add(Flatten())
model2.add(Dense(3, activation ='softmax'))
#Dataset


# Optimization
model2.compile(loss='categorical_crossentropy', metrics=["accuracy"],optimizer="adam")
model2.fit(x_train, y_train, batch_size=32, epochs=20);
```

```python
model2.summary()
```

```python
model2.evaluate(x_train, y_train)
```

### Summary :

Our previous linear model overfit because of the noise. It is not able to predict correctly.
Meanwhile using convolution allows us to learn even with noise. The convolution can detect shape and do not simply use the value of each pixel independently. The new model doesn't overfit with the noise and should be prefered.

### 5 A Regression Problem

```python
[X_train, Y_train] = generate_dataset_regression(3000, 20)
[X_test,Y_test] =generate_test_set_regression()
visualize_prediction(X_train[0,:], Y_train[0])
```

```python
print(Y_train[0])
```

```python
### Normalizer
def Normalizer(Y):
    ordo = Y[[1,3,5]]
    absc = Y[[0,2,4]]
    S = np.argsort(ordo)
    ordo = ordo[S]
    absc = absc[S]
    Y2=[]
    for i in range(3):
        Y2.append(absc[i])
        Y2.append(ordo[i])
    return(np.array(Y2))


for i in range(Y_train.shape[0]):
    Y_train[i] = Normalizer(Y_train[i,:])

for j in range(Y_test.shape[0]):
    Y_test[j] = Normalizer(Y_test[j])
```

```python
# MODEL DEFINITION :
from keras.layers import Dropout
model3 = Sequential()
model3.add(Conv2D(32, kernel_size=(5,5),padding="same", activation="relu"))
model3.add(MaxPooling2D(pool_size=(2,2)))
model3.add(Conv2D(16, kernel_size=(5,5),padding="same", activation="relu"))
model3.add(MaxPooling2D(pool_size=(2,2)))
model3.add(Conv2D(8, kernel_size=(5,5),padding="same", activation="relu"))
model3.add(MaxPooling2D(pool_size=(2,2)))
model3.add(Flatten())
model3.add(Dropout(0.4))
model3.add(Dense(6, activation="relu"))

x_train = X_train.reshape(X_train.shape[0], 72,72, 1)
y_train = Y_train.copy()

model3.compile(loss='mse', optimizer="adam")
model3.fit(x_train, y_train, batch_size=32, epochs=25, validation_split=0.01,verbose=1, shuffle=True);
```

```python
x_test = X_test.reshape(X_test.shape[0], 72,72, 1)
y_test = Y_test.copy()
print(y_test.shape)
for i in range(15):
    visualize_prediction(x_test[i], y_test[i,:])
```

```python
print("Test loss :",model3.evaluate(x_test, y_test, verbose=0))
print("Train loss :",model3.evaluate(x_train, y_train, verbose=0))
```
